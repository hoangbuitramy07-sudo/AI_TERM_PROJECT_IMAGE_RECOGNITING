```python
"""
=====================================================================
 HE THONG NHAN DIEN MON AN CANTEEN DAI HOC - TRAIN PURE CNN
=====================================================================


Muc tieu:
 1. Nhan anh crop tung mon an.
 2. Train CNN tu dau de nhan dien ten mon.
 3. So khop bang gia va tinh tong tien.
 4. Kiem tra data leakage.
 5. Danh gia Accuracy, Precision, Recall, F1-score, Confusion Matrix.


Rang buoc:
 - Chi dung CNN tu train tu dau.
 - Khong dung EfficientNet/ResNet/MobileNet/DenseNet/ViT.
 - Khong dung Transfer Learning hoac pretrained model.
 - YOLO khong dung de nhan dien mon an; chi duoc gan ngoai ham de dem trung
   trong mon thit_kho_trung neu do an yeu cau.
"""


from __future__ import annotations


import hashlib
import json
import os
import random
import unicodedata
from collections import Counter, defaultdict
from dataclasses import dataclass
from pathlib import Path
from typing import Callable


os.environ.setdefault("MPLCONFIGDIR", os.path.join(os.getcwd(), ".matplotlib_cache"))
os.environ.setdefault("TF_CPP_MIN_LOG_LEVEL", "2")


import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from PIL import Image, UnidentifiedImageError
from sklearn.metrics import (
   accuracy_score,
   classification_report,
   confusion_matrix,
   f1_score,
   precision_score,
   recall_score,
)
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from ultralytics import YOLO



# ---------------------------------------------------------------------
# 1. Cau hinh
# ---------------------------------------------------------------------
SEED = 42
DATASET_DIR = Path("/content/drive/MyDrive/Tray_food_2")
TRAIN_DIR = DATASET_DIR / "train"
VAL_DIR = DATASET_DIR / "val"
TEST_DIR = DATASET_DIR / "test"


# 192 giu du thong tin texture/mau sac hon 128, nhung van phu hop dataset nho.
IMG_SIZE = 192
BATCH_SIZE = 24
EPOCHS = 200
LEARNING_RATE = 3e-4
VALIDATION_SPLIT = 0.2
L2_WEIGHT = 2e-4
LABEL_SMOOTHING = 0.03


MODEL_DIR = Path("checkpoints")
BEST_MODEL_PATH = MODEL_DIR / "best_model.keras"
FINAL_MODEL_PATH = Path("canteen_cnn_final.keras")
CLASS_INDICES_PATH = Path("class_indices.json")


IMAGE_EXTENSIONS = {
   ".jpg", ".jpeg", ".png", ".bmp", ".gif", ".webp"
}


CANONICAL_CLASSES = [
   "com_trang",
   "dau_hu_sot_ca",
   "ca_hu_kho",
   "thit_kho_trung",
   "thit_kho",
   "canh_chua_co_ca",
   "canh_chua_khong_ca",
   "suon_nuong",
   "canh_rau_cai_thao",
   "canh_rau_muong",
   "rau_xao_cu_san",
   "rau_xao_dau_dua",
   "rau_xao_dau_que",
   "rau_xao_lagim",
   "trung_chien_thit",
   "trung_chien",
]


PRICE_TABLE = {
   "com_trang": 10_000,
   "dau_hu_sot_ca": 25_000,
   "ca_hu_kho": 30_000,
   "thit_kho_trung": 30_000,
   "thit_kho": 25_000,
   "canh_chua_co_ca": 25_000,
   "canh_chua_khong_ca": 10_000,
   "suon_nuong": 30_000,
   "canh_rau": 7_000,
   "rau_xao": 10_000,
   "trung_chien": 25_000,
}




def normalize_key(text: str) -> str:
   """Normalize Unicode folder/class names from macOS/Drive."""
   return unicodedata.normalize("NFC", text).strip().lower()




CLASS_ALIASES = {
   "com": "com_trang",
   "dau_hu_sot_ca": "dau_hu_sot_ca",
   "ca_hu_kho": "ca_hu_kho",
   "thit_kho_trung": "thit_kho_trung",
   "thit_kho": "thit_kho",
   "canh_chua_co_ca": "canh_chua_co_ca",
   "canh_chua_khong_ca": "canh_chua_khong_ca",
   "suon_nuong": "suon_nuong",
   "canh_rau_muong": "canh_rau_muong",
   "canh_rau_cai_thao": "canh_rau_cai_thao",
   "rau_xao_lagim": "rau_xao_lagim",
   "rau_xao_cu_san": "rau_xao_cu_san",
   "rau_xao_dau_que": "rau_xao_dau_que",
   "rau_xao_dau_dua": "rau_xao_dau_dua",
   "trung_chien": "trung_chien",
   "trung_chien_thit": "trung_chien_thit",
}
CLASS_ALIASES = {normalize_key(k): v for k, v in CLASS_ALIASES.items()}




@dataclass(frozen=True)
class ImageSample:
   path: Path
   label: str
   source_folder: str
   content_hash: str




# ---------------------------------------------------------------------
# 2. Dataset, alias 11 lop, va leakage check
# ---------------------------------------------------------------------
def set_reproducibility(seed: int = SEED) -> None:
   random.seed(seed)
   np.random.seed(seed)
   tf.random.set_seed(seed)




def sha256_file(path: Path) -> str:
   digest = hashlib.sha256()
   with path.open("rb") as f:
       for chunk in iter(lambda: f.read(1024 * 1024), b""):
           digest.update(chunk)
   return digest.hexdigest()




def is_valid_image_file(path: Path) -> bool:
   """Skip corrupted/unsupported images before TensorFlow starts training."""
   try:
       with Image.open(path) as img:
           img.verify()
       return True
   except (OSError, UnidentifiedImageError):
       return False




def canonical_label_from_folder(folder_name: str) -> str | None:
   return CLASS_ALIASES.get(normalize_key(folder_name))




def scan_split(split_dir: Path) -> list[ImageSample]:
   if not split_dir.exists():
       return []


   samples: list[ImageSample] = []
   unknown_folders: list[str] = []
   skipped_files: list[Path] = []
   for class_dir in sorted(p for p in split_dir.iterdir() if p.is_dir()):
       canonical = canonical_label_from_folder(class_dir.name)
       if canonical is None:
           unknown_folders.append(class_dir.name)
           continue


       for img_path in sorted(class_dir.rglob("*")):
           if not img_path.is_file():
               continue
           if img_path.suffix.lower() not in IMAGE_EXTENSIONS:
               continue
           if not is_valid_image_file(img_path):
               skipped_files.append(img_path)
               continue


           samples.append(
               ImageSample(
                   path=img_path,
                   label=canonical,
                   source_folder=class_dir.name,
                   content_hash=sha256_file(img_path),
               )
           )


   if unknown_folders:
       print("Canh bao: bo qua folder chua map ve 11 lop:", unknown_folders)
   if skipped_files:
       print(f"Canh bao: bo qua {len(skipped_files)} file anh hong/khong doc duoc.")
       for bad_path in skipped_files[:10]:
           print(f"  - {bad_path}")
   return samples




def deduplicate_exact_images(samples: list[ImageSample]) -> list[ImageSample]:
   """Keep one copy per (content hash, canonical label)."""
   seen: set[tuple[str, str]] = set()
   unique_samples: list[ImageSample] = []
   for sample in samples:
       key = (sample.content_hash, sample.label)
       if key not in seen:
           seen.add(key)
           unique_samples.append(sample)
   removed = len(samples) - len(unique_samples)
   if removed:
       print(f"Da loai {removed} anh trung lap y het trong cung tap du lieu.")
   return unique_samples




def print_distribution(name: str, samples: list[ImageSample]) -> None:
   counts = Counter(sample.label for sample in samples)
   print(f"\n{name}: {len(samples)} anh")
   for class_name in CANONICAL_CLASSES:
       if counts[class_name] > 0:
           print(f"  {class_name:<22}: {counts[class_name]}")




def split_train_validation_by_hash(
   samples: list[ImageSample],
   validation_split: float = VALIDATION_SPLIT,
) -> tuple[list[ImageSample], list[ImageSample]]:
   """
   Split theo content hash de anh duplicate khong roi vao ca train va val.
   Moi hash lay label cua sample dau tien; neu 1 hash co nhieu label thi bao loi.
   """
   by_hash: dict[str, list[ImageSample]] = defaultdict(list)
   for sample in samples:
       by_hash[sample.content_hash].append(sample)


   inconsistent = []
   for content_hash, group in by_hash.items():
       labels = {sample.label for sample in group}
       if len(labels) > 1:
           inconsistent.append((content_hash, sorted(labels), [str(s.path) for s in group]))
   if inconsistent:
       print("\nLOI: cung mot anh nhung nam o nhieu label khac nhau.")
       for _, labels, paths in inconsistent[:10]:
           print("  labels=", labels, "paths=", paths)
       raise ValueError("Dataset co duplicate conflict label. Hay sua truoc khi train.")


   hashes = list(by_hash.keys())
   labels = [by_hash[h][0].label for h in hashes]
   label_counts = Counter(labels)
   stratify = labels if min(label_counts.values()) >= 2 else None


   train_hashes, val_hashes = train_test_split(
       hashes,
       test_size=validation_split,
       random_state=SEED,
       shuffle=True,
       stratify=stratify,
   )
   train_hashes = set(train_hashes)
   val_hashes = set(val_hashes)


   train_samples = [s for h in train_hashes for s in by_hash[h]]
   val_samples = [s for h in val_hashes for s in by_hash[h]]
   return train_samples, val_samples




def check_data_leakage(
   train_samples: list[ImageSample],
   val_samples: list[ImageSample],
   test_samples: list[ImageSample],
) -> dict[str, list[tuple[str, list[str]]]]:
   split_map = {
       "train": train_samples,
       "val": val_samples,
       "test": test_samples,
   }
   hash_to_locations: dict[str, list[tuple[str, ImageSample]]] = defaultdict(list)
   for split_name, samples in split_map.items():
       for sample in samples:
           hash_to_locations[sample.content_hash].append((split_name, sample))


   leaks: dict[str, list[tuple[str, list[str]]]] = {}
   pairs = [("train", "val"), ("train", "test"), ("val", "test")]
   for left, right in pairs:
       pair_key = f"{left}_vs_{right}"
       pair_leaks = []
       for content_hash, locations in hash_to_locations.items():
           split_names = {split_name for split_name, _ in locations}
           if left in split_names and right in split_names:
               paths = [
                   f"{split_name}: {sample.label}: {sample.path}"
                   for split_name, sample in locations
                   if split_name in {left, right}
               ]
               pair_leaks.append((content_hash, paths))
       leaks[pair_key] = pair_leaks


   print("\n===== DATA LEAKAGE CHECK =====")
   for pair_key, pair_leaks in leaks.items():
       print(f"{pair_key}: {len(pair_leaks)} duplicate hash")
       for _, paths in pair_leaks[:5]:
           print("  " + " | ".join(paths))
   return leaks




def prepare_datasets() -> tuple[list[ImageSample], list[ImageSample], list[ImageSample], list[str]]:
   raw_train = deduplicate_exact_images(scan_split(TRAIN_DIR))
   raw_val = deduplicate_exact_images(scan_split(VAL_DIR))
   raw_test = deduplicate_exact_images(scan_split(TEST_DIR))


   if not raw_train:
       raise FileNotFoundError(f"Khong tim thay anh train trong {TRAIN_DIR}")


   if raw_val:
       train_samples, val_samples = raw_train, raw_val
   else:
       print(f"\nKhong co anh trong {VAL_DIR}. Tu tach {VALIDATION_SPLIT:.0%} tu train theo hash.")
       train_samples, val_samples = split_train_validation_by_hash(raw_train)
   test_samples = raw_test


   active_classes = [
       class_name for class_name in CANONICAL_CLASSES
       if any(sample.label == class_name for sample in train_samples + val_samples + test_samples)
   ]
   missing_classes = [c for c in CANONICAL_CLASSES if c not in active_classes]
   if missing_classes:
       print("\nCanh bao: dataset hien tai chua co anh cho lop:", missing_classes)


   print_distribution("Train", train_samples)
   print_distribution("Validation", val_samples)
   if test_samples:
       print_distribution("Test", test_samples)
   else:
       print("\nTest: 0 anh. Se bo qua danh gia test rieng.")


   check_data_leakage(train_samples, val_samples, test_samples)
   return train_samples, val_samples, test_samples, active_classes




# ---------------------------------------------------------------------
# 3. tf.data pipeline
# ---------------------------------------------------------------------
def make_class_maps(class_names: list[str]) -> tuple[dict[str, int], dict[int, str]]:
   class_indices = {class_name: idx for idx, class_name in enumerate(class_names)}
   idx_to_class = {idx: class_name for class_name, idx in class_indices.items()}
   with CLASS_INDICES_PATH.open("w", encoding="utf-8") as f:
       json.dump(class_indices, f, ensure_ascii=False, indent=2)
   print(f"\nDa luu {CLASS_INDICES_PATH}")
   return class_indices, idx_to_class




def samples_to_arrays(
   samples: list[ImageSample],
   class_indices: dict[str, int],
) -> tuple[np.ndarray, np.ndarray]:
   paths = np.array([str(sample.path) for sample in samples])
   labels = np.array([class_indices[sample.label] for sample in samples], dtype=np.int32)
   return paths, labels




def decode_resize(path: tf.Tensor, label: tf.Tensor, num_classes: int) -> tuple[tf.Tensor, tf.Tensor]:
   image_bytes = tf.io.read_file(path)
   image = tf.io.decode_image(image_bytes, channels=3, expand_animations=False)
   image.set_shape([None, None, 3])
   image = tf.image.resize(image, [IMG_SIZE, IMG_SIZE], method="bilinear")
   image = tf.cast(image, tf.float32)
   label = tf.one_hot(label, depth=num_classes)
   return image, label




def build_tf_dataset(
   samples: list[ImageSample],
   class_indices: dict[str, int],
   training: bool,
) -> tf.data.Dataset:
   paths, labels = samples_to_arrays(samples, class_indices)
   ds = tf.data.Dataset.from_tensor_slices((paths, labels))
   if training:
       ds = ds.shuffle(buffer_size=len(samples), seed=SEED, reshuffle_each_iteration=True)
   ds = ds.map(
       lambda p, y: decode_resize(p, y, len(class_indices)),
       num_parallel_calls=tf.data.AUTOTUNE,
   )
   ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
   return ds




# ---------------------------------------------------------------------
# 4. CNN thuan: Conv2D -> BatchNorm -> MaxPooling -> Dense -> Softmax
# ---------------------------------------------------------------------
def conv_block(x: tf.Tensor, filters: int, dropout: float) -> tf.Tensor:
   for _ in range(2):
       x = layers.Conv2D(
           filters,
           kernel_size=3,
           padding="same",
           use_bias=False,
           kernel_initializer="he_normal",
           kernel_regularizer=regularizers.l2(L2_WEIGHT),
       )(x)
       x = layers.BatchNormalization()(x)
       x = layers.Activation("relu")(x)
   x = layers.MaxPooling2D(pool_size=2)(x)
   if dropout > 0:
       x = layers.Dropout(dropout)(x)
   return x




def build_pure_cnn(num_classes: int, input_shape: tuple[int, int, int] = (IMG_SIZE, IMG_SIZE, 3)) -> keras.Model:
   inputs = layers.Input(shape=input_shape)


   data_augmentation = keras.Sequential(
       [
           layers.RandomFlip("horizontal"),
           layers.RandomRotation(0.08),
           layers.RandomTranslation(height_factor=0.08, width_factor=0.08),
           layers.RandomZoom(height_factor=(-0.12, 0.12), width_factor=(-0.12, 0.12)),
           layers.RandomContrast(0.18),
       ],
       name="data_augmentation",
   )


   x = data_augmentation(inputs)
   x = layers.Rescaling(1.0 / 255.0)(x)


   x = conv_block(x, 32, dropout=0.05)
   x = conv_block(x, 64, dropout=0.08)
   x = conv_block(x, 128, dropout=0.12)
   x = conv_block(x, 256, dropout=0.18)
   x = conv_block(x, 384, dropout=0.22)


   x = layers.GlobalAveragePooling2D()(x)
   x = layers.Dense(
       512,
       activation="relu",
       kernel_regularizer=regularizers.l2(L2_WEIGHT),
   )(x)
   x = layers.BatchNormalization()(x)
   x = layers.Dropout(0.45)(x)
   x = layers.Dense(
       128,
       activation="relu",
       kernel_regularizer=regularizers.l2(L2_WEIGHT),
   )(x)
   x = layers.Dropout(0.25)(x)
   outputs = layers.Dense(num_classes, activation="softmax")(x)


   return keras.Model(inputs=inputs, outputs=outputs, name="canteen_pure_cnn")




def compile_model(model: keras.Model) -> None:
   model.compile(
       optimizer=keras.optimizers.Adam(learning_rate=LEARNING_RATE, clipnorm=1.0),
       loss=keras.losses.CategoricalCrossentropy(label_smoothing=LABEL_SMOOTHING),
       metrics=["accuracy"],
   )




def make_callbacks() -> list[keras.callbacks.Callback]:
   MODEL_DIR.mkdir(exist_ok=True)
   return [
       keras.callbacks.ModelCheckpoint(
           filepath=BEST_MODEL_PATH,
           monitor="val_accuracy",
           mode="max",
           save_best_only=True,
           verbose=1,
       ),
       keras.callbacks.EarlyStopping(
           monitor="val_loss",
           mode="min",
           patience=15,
           restore_best_weights=True,
           verbose=1,
       ),
       keras.callbacks.ReduceLROnPlateau(
           monitor="val_loss",
           mode="min",
           factor=0.3,
           patience=5,
           cooldown=1,
           min_lr=1e-6,
           verbose=1,
       ),
   ]




def compute_class_weights(samples: list[ImageSample], class_indices: dict[str, int]) -> dict[int, float]:
   y = np.array([class_indices[sample.label] for sample in samples], dtype=np.int32)
   classes = np.arange(len(class_indices), dtype=np.int32)
   weights = compute_class_weight(class_weight="balanced", classes=classes, y=y)
   return {int(class_id): float(weight) for class_id, weight in zip(classes, weights)}




# ---------------------------------------------------------------------
# 5. Bao cao metric, confusion matrix, history
# ---------------------------------------------------------------------
def plot_history(history: keras.callbacks.History) -> None:
   fig, axes = plt.subplots(1, 2, figsize=(14, 5))


   axes[0].plot(history.history["accuracy"], label="Train Acc")
   axes[0].plot(history.history["val_accuracy"], label="Val Acc")
   axes[0].set_title("Accuracy")
   axes[0].set_xlabel("Epoch")
   axes[0].legend()
   axes[0].grid(True)


   axes[1].plot(history.history["loss"], label="Train Loss")
   axes[1].plot(history.history["val_loss"], label="Val Loss")
   axes[1].set_title("Loss")
   axes[1].set_xlabel("Epoch")
   axes[1].legend()
   axes[1].grid(True)


   plt.tight_layout()
   plt.savefig("training_history.png", dpi=150)
   plt.show()
   print("Da luu bieu do: training_history.png")




def predict_dataset(model: keras.Model, ds: tf.data.Dataset) -> tuple[np.ndarray, np.ndarray]:
   y_true_batches = []
   y_pred_batches = []
   for images, labels in ds:
       probs = model.predict(images, verbose=0)
       y_pred_batches.append(np.argmax(probs, axis=1))
       y_true_batches.append(np.argmax(labels.numpy(), axis=1))
   return np.concatenate(y_true_batches), np.concatenate(y_pred_batches)




def evaluate_and_report(
   model: keras.Model,
   ds: tf.data.Dataset,
   class_names: list[str],
   split_name: str,
) -> dict[str, float]:
   loss, acc = model.evaluate(ds, verbose=1)
   y_true, y_pred = predict_dataset(model, ds)


   report = classification_report(
       y_true,
       y_pred,
       target_names=class_names,
       zero_division=0,
       digits=4,
   )
   print(f"\n===== CLASSIFICATION REPORT: {split_name.upper()} =====")
   print(report)


   metrics = {
       "loss": float(loss),
       "accuracy": float(accuracy_score(y_true, y_pred)),
       "keras_accuracy": float(acc),
       "precision_weighted": float(precision_score(y_true, y_pred, average="weighted", zero_division=0)),
       "recall_weighted": float(recall_score(y_true, y_pred, average="weighted", zero_division=0)),
       "f1_weighted": float(f1_score(y_true, y_pred, average="weighted", zero_division=0)),
       "precision_macro": float(precision_score(y_true, y_pred, average="macro", zero_division=0)),
       "recall_macro": float(recall_score(y_true, y_pred, average="macro", zero_division=0)),
       "f1_macro": float(f1_score(y_true, y_pred, average="macro", zero_division=0)),
   }


   print(f"\n===== MODEL PERFORMANCE: {split_name.upper()} =====")
   for key, value in metrics.items():
       print(f"{key:<20}: {value:.4f}")


   with open(f"{split_name}_classification_report.txt", "w", encoding="utf-8") as f:
       f.write(report)
   with open(f"{split_name}_metrics.json", "w", encoding="utf-8") as f:
       json.dump(metrics, f, ensure_ascii=False, indent=2)


   cm = confusion_matrix(y_true, y_pred, labels=np.arange(len(class_names)))
   plt.figure(figsize=(12, 10))
   plt.imshow(cm, interpolation="nearest", cmap="Blues")
   plt.title(f"Confusion Matrix - {split_name}")
   plt.colorbar()
   tick_marks = np.arange(len(class_names))
   plt.xticks(tick_marks, class_names, rotation=45, ha="right")
   plt.yticks(tick_marks, class_names)


   thresh = cm.max() / 2.0 if cm.max() else 0
   for i in range(cm.shape[0]):
       for j in range(cm.shape[1]):
           plt.text(
               j,
               i,
               format(cm[i, j], "d"),
               ha="center",
               va="center",
               color="white" if cm[i, j] > thresh else "black",
           )
   plt.ylabel("True")
   plt.xlabel("Predicted")
   plt.tight_layout()
   plt.savefig(f"{split_name}_confusion_matrix.png", dpi=150)
   plt.show()
   print(f"Da luu confusion matrix: {split_name}_confusion_matrix.png")
   return metrics




# ---------------------------------------------------------------------
# 6. Nhan dien mon da crop va tinh tien
# --- YOLO dem trung ---
_yolo_model: YOLO | None = None

def get_yolo_model() -> YOLO:
    global _yolo_model
    if _yolo_model is None:
        _yolo_model = YOLO("/content/drive/MyDrive/best.pt")
        print("Da load YOLO model: best.pt")
    return _yolo_model

def count_eggs_yolo(img_path: str | Path) -> int:
    results = get_yolo_model()(str(img_path), verbose=False)
    return max(1, int(len(results[0].boxes)))

def load_class_indices() -> tuple[dict[str, int], dict[int, str]]:
   with CLASS_INDICES_PATH.open("r", encoding="utf-8") as f:
       class_indices = json.load(f)
   idx_to_class = {int(v): k for k, v in class_indices.items()}
   return class_indices, idx_to_class

def predict_single_image(
   model: keras.Model,
   img_path: str | Path,
   idx_to_class: dict[int, str],
) -> tuple[str, float]:
   img = keras.utils.load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
   img_array = keras.utils.img_to_array(img)
   img_batch = np.expand_dims(img_array, axis=0)
   preds = model.predict(img_batch, verbose=0)[0]
   class_idx = int(np.argmax(preds))
   return idx_to_class[class_idx], float(preds[class_idx])




def predict_tray(
   image_paths: list[str | Path],
   model_path: str | Path = BEST_MODEL_PATH,
   egg_counter: Callable[[str | Path], int] | None = None,
) -> None:
   """
   Nhan danh sach anh da crop, nhan dien bang CNN va tinh tien.


   egg_counter chi duoc dung cho thit_kho_trung. Neu ban co YOLO dem trung,
   truyen vao ham dang: egg_counter(image_path) -> so_luong_trung.
   CNN van la model duy nhat nhan dien ten mon.
   """
   loaded_model = keras.models.load_model(model_path)
   _, idx_to_class = load_class_indices()


   print("\n" + "=" * 72)
   print("  KET QUA NHAN DIEN MON AN & TINH TIEN")
   print("=" * 72)
   print(f"{'STT':<5} {'Ten mon':<24} {'Gia':>12} {'Ghi chu':<20}")
   print("-" * 72)


   total = 0
   for idx, img_path in enumerate(image_paths, start=1):
       dish_name, confidence = predict_single_image(loaded_model, img_path, idx_to_class)
       price = PRICE_TABLE.get(dish_name, 0)
       note = f"conf={confidence:.2f}"


       if dish_name == "thit_kho_trung" and egg_counter is not None:
           egg_count = max(1, int(egg_counter(img_path)))
           extra_eggs = max(0, egg_count - 1)
           price = PRICE_TABLE[dish_name] + extra_eggs * 6_000
           note += f", trung={egg_count}"


       total += price
       print(f"{idx:<5} {dish_name:<24} {price:>10,}d  {note:<20}")


   print("-" * 72)
   print(f"{'TONG TIEN':<31} {total:>10,}d")
   print("=" * 72)




def demo_crop_and_predict(
   tray_image_path: str | Path,
   crop_coords: list[tuple[int, int, int, int]],
   output_dir: str | Path = "cropped_dishes",
) -> None:
   output_dir = Path(output_dir)
   output_dir.mkdir(exist_ok=True)
   img = plt.imread(tray_image_path)


   saved_paths = []
   for i, (x, y, w, h) in enumerate(crop_coords, start=1):
       cropped = img[y : y + h, x : x + w]
       save_path = output_dir / f"dish_{i}.jpg"
       plt.imsave(save_path, cropped)
       saved_paths.append(save_path)
       print(f"Da luu anh crop: {save_path}")


   predict_tray(saved_paths)




# ---------------------------------------------------------------------
# 7. Main training
# ---------------------------------------------------------------------
def main() -> None:
   set_reproducibility()
   print("TensorFlow version:", tf.__version__)
   print("GPU available:", tf.config.list_physical_devices("GPU"))


   train_samples, val_samples, test_samples, class_names = prepare_datasets()
   class_indices, _ = make_class_maps(class_names)


   train_ds = build_tf_dataset(train_samples, class_indices, training=True)
   val_ds = build_tf_dataset(val_samples, class_indices, training=False)
   test_ds = build_tf_dataset(test_samples, class_indices, training=False) if test_samples else None


   model = build_pure_cnn(num_classes=len(class_names))
   compile_model(model)
   model.summary()


   class_weight = compute_class_weights(train_samples, class_indices)
   print("\nClass weight:", class_weight)


   print("\n========== BAT DAU TRAIN PURE CNN ==========")
   history = model.fit(
       train_ds,
       epochs=EPOCHS,
       validation_data=val_ds,
       callbacks=make_callbacks(),
       class_weight=class_weight,
       verbose=1,
   )


   model.save(FINAL_MODEL_PATH)
   print(f"Da luu model cuoi: {FINAL_MODEL_PATH}")
   plot_history(history)


   best_model = keras.models.load_model(BEST_MODEL_PATH)
   evaluate_and_report(best_model, val_ds, class_names, split_name="val")


   if test_ds is not None:
       evaluate_and_report(best_model, test_ds, class_names, split_name="test")
   else:
       print("\nKhong co test set. Bao cao tren validation chi dung de tham khao,")
       print("nen tach test doc lap truoc khi ket luan accuracy cuoi cung.")


   print("\nUoc luong sau toi uu:")
   print("- Neu crop on dinh, anh ro va split khong leakage: validation co the tang len khoang 70-85%.")
   print("- Neu anh test khac goc chup/anh sang/khay: ky vong thuc te nen bao thu hon, khoang 65-78%.")




if __name__ == "__main__":
   main()
```
